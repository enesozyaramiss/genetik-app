```python
import keyboard
import time
from datetime import datetime
import os

while True:
    # Şu anki saat ve dakika
    now = datetime.now()
    current_hour = now.hour
    current_minute = now.minute

    # Eğer saat 17:59 veya sonrası ise bilgisayarı kapat
    if current_hour > 17 or (current_hour == 17 and current_minute >= 59):
        os.system("shutdown /s /t 1")
        break

    # SPACE tuşuna bas  
    keyboard.press_and_release('space')
    print(f"Pressed SPACE at {now.strftime('%H:%M:%S')}")

    # 5 dakika bekle (300 saniye)
    time.sleep(240)

print("Program finished.")
```

```python
import pandas as pd
import gzip
def parse_vcf_gz_to_df(file_path):
    with gzip.open(file_path, 'rt') as f:
        lines = [l for l in f if not l.startswith('##')]  # Sadece header ve veri satırları
    header = lines[0].strip().lstrip('#').split('\t')     # VCF sütun isimleri
    data = [line.strip().split('\t') for line in lines[1:]]
    return pd.DataFrame(data, columns=header)
```

```python
import requests
def get_pubmed_ids_from_clinvar(variation_id):
    url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi"
    params = {
        "dbfrom": "clinvar",
        "db": "pubmed",
        "id": variation_id,
        "retmode": "json"
    }
    response = requests.get(url, params=params)
    data = response.json()
    linksets = data.get("linksets", [])
    if not linksets or "linksetdbs" not in linksets[0]:
        return []
    pmids = []
    for db in linksets[0]["linksetdbs"]:
        if db["dbto"] == "pubmed":
            pmids.extend(db["links"])
    return pmids
```

```python
print(get_pubmed_ids_from_clinvar(370969))
```

```python
def get_pubmed_links(pmid_list):
    return [f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/" for pmid in pmid_list]

# Örnek kullanım:
pmids = get_pubmed_ids_from_clinvar(3385321)
links = get_pubmed_links(pmids)
print(links)
```
